```python
import numpy as np
import pandas as pd
import re
import string
import pickle
import joblib
```

Load saved modles

```python
heart_model = joblib.load("../artifacts/heart_disease_model.pkl")
diabetes_model = joblib.load("../artifacts/diabetes_model.pkl")
scaler = joblib.load("../artifacts/scaler.pkl")
feature_names = joblib.load("../artifacts/feature_names.pkl")
```

Create prediction Function

```python
def predict_risk(user_input):
    """
    user_input: list of values in SAME order as training features
    """

    user_input_df = pd.DataFrame([user_input], columns=feature_names)

    # scale (only if you used scaled features)
    user_input_scaled = scaler.transform(user_input_df)

    heart_pred = heart_model.predict(user_input_df)[0]
    diabetes_pred = diabetes_model.predict(user_input_df)[0]

    return {
        "Heart Disease Risk": "High" if heart_pred == 1 else "Low",
        "Diabetes Risk": "High" if diabetes_pred == 1 else "Low"
    }
```
